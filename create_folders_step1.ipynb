```python
import os
import nibabel as nib
import glob
from PIL import Image
import numpy as np
import pandas as pd
import shutil
```

```python
folder_path = '/data/scratch/chirag24/HIA'
ff = glob.glob(folder_path+'/cropped/*')
ff[0]
```

```python
images = []
```

```python
for f in range(len(ff)):
    try:
        folderName1 = ff[f].split("cropped/")[1].split(".")[0]
        folderName = folder_path+"/data/"+folderName1
        if not os.path.exists(folderName):
            os.mkdir(folderName)
        a = nib.load(ff[f])
        a = a.get_fdata()
        a = a[:,:,:]
        for i in range(a.shape[1]):
            name = folderName+"/"+folderName1+"-s"+str(i)+".png"
            #im = Image.fromarray(a[:,i,:])
            im = a[:,i,:]
            im = (im-np.min(im))/np.max(im)*255
            im = Image.fromarray(im[:,:])
            if im.mode != 'RGB':
                im = im.convert('RGB')
            im.save(name)
    except:
        print("Error")
```

```python
df = pd.read_csv(folder_path+"/HR-MICRA_rawdata_export.csv", header=None)
df.head()
```

```python
columns = list(df) 
```

```python
cnt = 0
for i in columns:
    if cnt%2 == 0 and i!=0:
        parent_folder = df[i][0]
        if str(parent_folder) == "nan":
            break
        parent_folder = folder_path+"/data/"+parent_folder.zfill(2)
        if not os.path.exists(parent_folder):
            os.mkdir(str(parent_folder))
        sub_folder =  str(df[i][5]).zfill(2)
        left =  sub_folder+"cl"
        right = sub_folder+"cr"
        try:
            shutil.move(folder_path+"/data/"+left, parent_folder+"/"+left) 
            shutil.move(folder_path+"/data/"+right, parent_folder+"/"+right)
        except:
            print(folder_path+"/data/"+left)
        
    cnt = cnt +1
```
